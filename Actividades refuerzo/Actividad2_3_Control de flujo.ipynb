**1.**-  Escribe in programa que pida un número al usuario y muestre su tabla de multiplicar. El número debe ser del 1 al 9. Debes hacerlo con una sentencia while.

```python
# escribe tu código

numero = int(input("Dame un numero: "))
while numero < 0 or numero > 9:
    print("Solo puedes entre 1 y 9")
    numero = int(input("Dame otro numero: "))
for numeros in range(11):
    calculo = numeros * numero
    print(calculo)
```
