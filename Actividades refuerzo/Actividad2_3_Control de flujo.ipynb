**1.**-  Escribe in programa que pida un número al usuario y muestre su tabla de multiplicar. El número debe ser del 1 al 9. Debes hacerlo con una sentencia while.

```python
# escribe tu código

numero = int(input("Dame un numero: "))
while numero < 0 or numero > 9:
    print("Solo puedes entre 1 y 9")
    numero = int(input("Dame otro numero: "))
for numeros in range(11):
    calculo = numeros * numero
    print(calculo)
```

**3.**- 
Escribe un programa que pida números enteros de uno en uno, los meta en una lista y, cuando acabe (debes pedir introducir una letra en concreto para acabar) debe recorrer la lista y mostrar la suma de los pares y la suma de los impares.

```python
# escribe el código

numeros = []
suma_pares = 0
suma_impares = 0

numero = input('Introduce un numero o escriba "f" para acabar: ')
while numero != "f" or type(numero) == int():
    numeros.append(numero)
    numero = input('Introduce otro numero o escriba "f" para acabar: ')
for cifra in numeros:
    if int(cifra) % 2 == 0:
        suma_pares = suma_pares + int(cifra)
    elif int(cifra) % 2 != 0:
        suma_impares = suma_impares + int(cifra)

print("La suma de pares es " + str(suma_pares) + " " + "y la de impares es " + str(suma_impares))
```

**4.-** Escribe un programa que muestre un menú (debes hacerlo con un while) como este:

1) crea una lista nueva.

2) añade elemento a la lista.

3) muestra los elementos de tipo numérico.

4) muestra solo los elementos pares.

5) salir.


En la primera opción debe crear una lista nueva. Es decir, si ya se había creado una y se habían añadido elementos, se crea una nueva vacía y se pierde.
En la segunda opción (si el usuario pulsa un 2) debe solicitar un nuevo elemento y lo mete en la lista. Si el elemento es un número debe meterlo como tal, sino, como string.
En la tercera opción el programa debe recorrer la lista y mostrar sólo los elementos de tipo numérico.
En la cuarta solo los números que sean pares.
Si el usuario introduce un 5 el programa debe terminar.

```python
# escribe el código

print("""
    1) crea una lista nueva.
      
    2) añade elemento a la lista.

    3) muestra los elementos de tipo numérico.

    4) muestra solo los elementos pares.

    5) salir.""")

decision = input("Escoge una opcion: ")

while decision != "5":  # Comparar con string "5"

    if decision == "1":  # Comparar con string "1"
        lista = []
        print("Lista creada.")

    elif decision == "2":
        elemento = input("Añade algo: ")
        # Intentar convertir a número si es posible
        try:
            elemento = int(elemento)
        except ValueError:
            try:
                elemento = float(elemento)
            except ValueError:
                pass  # Mantener como string
        lista.append(elemento)
        print(f"Elemento '{elemento}' añadido.")

    elif decision == "3":
        print("Elementos numéricos:")
        for elemento in lista:
            if type(elemento) in [int, float]:  # Sin paréntesis
                print(elemento)

    elif decision == "4":
        print("Elementos pares:")
        for elemento in lista:
            if type(elemento) == int and elemento % 2 == 0:
                print(elemento)

    else:
        print("Opción no válida.")

    # Volver a mostrar el menú y pedir nueva opción
    print("""
    1) crea una lista nueva.
    2) añade elemento a la lista.
    3) muestra los elementos de tipo numérico.
    4) muestra solo los elementos pares.
    5) salir.""")
    decision = input("Escoge una opcion: ")

print("¡Hasta luego!")
                
```
