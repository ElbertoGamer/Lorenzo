1.2 Haz una función que intercambie dos elementos de una lista y recibiendo como parámetros la lista y las posiciones.

```python
# Función intercambia
def Intercambia(list, pos1, pos2):
     
    l1 = list[pos1]
    l2 = list[pos2]
    list.remove(l1)
    list.remove(l2)
    list.insert(pos1,l2)
    list.insert(pos2,l1) 

    #tu codigo aquí
    return list
 
# Probamos la función
List = [10, 12, 23, 71]
pos1, pos2  = 1, 3
print(Intercambia(List,pos1,pos2))
```
